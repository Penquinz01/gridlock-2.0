```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix
)

import joblib
```

```python
df.head()
```

```python
df=pd.read_csv("preprocessed_data.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df["requires_road_closure"].value_counts()
```

```python
X_rc = df[
    [
        "event_type",
        "event_cause",
        "veh_type",
        "hour",
        "day_of_week",
        "month"
    ]
]
```

```python
X_rc.head()
```

```python
y_rc = df["requires_road_closure"]
```

```python
y_rc.head()
```

```python
print("Features Shape:", X_rc.shape)
print("Target Shape:", y_rc.shape)
```

```python
X_train_rc, X_test_rc, y_train_rc, y_test_rc = train_test_split(
    X_rc,
    y_rc,
    test_size=0.2,
    random_state=42,
    stratify=y_rc
)
```

```python
print(X_train_rc.shape)
print(X_test_rc.shape)
```

```python
road_closure_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)
```

```python
road_closure_model.fit(
    X_train_rc,
    y_train_rc
)
```

```python
y_pred_rc = road_closure_model.predict(
    X_test_rc
)
```

```python
accuracy_rc = accuracy_score(
    y_test_rc,
    y_pred_rc
)

print("Road Closure Accuracy:", accuracy_rc)
```

```python
print(
    classification_report(
        y_test_rc,
        y_pred_rc
    )
)
```

```python
road_closure_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight="balanced"
)
```

```python
road_closure_model.fit(
    X_train_rc,
    y_train_rc
)
```

```python
y_pred_rc = road_closure_model.predict(
    X_test_rc
)
```

```python
print(
    classification_report(
        y_test_rc,
        y_pred_rc
    )
)
```

```python
from sklearn.metrics import confusion_matrix

cm_rc = confusion_matrix(
    y_test_rc,
    y_pred_rc
)

print(cm_rc)
```

```python
accuracy_rc = accuracy_score(
    y_test_rc,
    y_pred_rc
)

print("Road Closure Accuracy:", accuracy_rc)
```

```python
importance_rc = pd.DataFrame({
    "Feature": X_rc.columns,
    "Importance": road_closure_model.feature_importances_
})

importance_rc = importance_rc.sort_values(
    by="Importance",
    ascending=False
)

print(importance_rc)
```

```python
joblib.dump(
    road_closure_model,
    "road_closure_model.pkl"
)
```
